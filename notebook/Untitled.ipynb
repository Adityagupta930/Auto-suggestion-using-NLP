```python
import numpy as np
import pandas as pd
import textdistance
import re
```

```python
word = []

with open('autocorrect book.txt', 'r', encoding='utf-8') as f:
    data = f.read()
    data = data.lower()
    words = re.findall('\w+', data)
    words += words

print(words[0:10])
```

```python
len(words)
```

```python
V= set(words)
print(len(V))
```

```python
from collections import Counter
```

```python
words_freq_dict = {}
words_freq_dict = Counter(words)
```

```python
words_freq_dict.most_common()
```

```python
Total = sum(words_freq_dict.values())
probs = {}

for k in words_freq_dict.keys():
    probs[k] = words_freq_dict[k] / Total
```

```python
# def autocorrect(word):
#     word = word.lower()
#     if word in unique_words :
#         return('Your word seems to be correct', word)
#     else:
#         similarities = [1-(textdistance.Jaccard(qval=2).distance(v,word))  for v in word_freq.keys()]
#         df = pd.DataFrame.from_dict(probs, orient='index').reset_index()
#         df = df.rename(columns={'index':'Word',0:'Prob'})
#         df['Similarity'] = similarities
#         output = df.sort_values(['Similarity','Prob'],ascending=False).head(3)
#         return(output)
# autocorrect("helo")
def jaccard_distance(s1, s2):
    set1 = set(s1)
    set2 = set(s2)
    return 1 - len(set1.intersection(set2)) / len(set1.union(set2))

def autocorrect(word):
    word = word.lower()
    if word in V:
        return ('Your word seems to be correct', word)
    else:
        similarities = [1 - jaccard_distance(v, word) for v in words_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs, orient='index').reset_index()
        df = df.rename(columns={'index': 'Word', 0: 'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity', 'Prob'], ascending=False).head(3)
        return output
```
